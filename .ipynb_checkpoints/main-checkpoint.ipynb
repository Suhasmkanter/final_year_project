```python
# --- TEST DATA LOADER ---
def get_test_data(test_dir='./Testing'):
    test_paths = []
    test_labels = []
    
    for class_name in classes:  # same order as training
        class_dir = os.path.join(test_dir, class_name)
        for image_file in os.listdir(class_dir):
            if image_file.endswith(('.png', '.jpg', '.jpeg')):
                test_paths.append(os.path.join(class_dir, image_file))
                test_labels.append(class_name)
                
    test_images = open_images(test_paths)
    test_labels = encode_label(test_labels)
    
    return np.array(test_images), np.array(test_labels)

# --- LOAD TEST DATA ---
test_images, test_labels = get_test_data()
# --- EVALUATE ---
loss, accuracy = model.evaluate(test_images, test_labels)
print(f'Test Loss: {loss:.4f}')
print(f'Test Accuracy: {accuracy*100:.2f}%')
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import pandas as pd

# --- PREDICT ---
preds = model.predict(test_images)
pred_classes = np.argmax(preds, axis=1)

# --- CLASSIFICATION REPORT ---
print(classification_report(test_labels, pred_classes, target_names=classes))

# --- CONFUSION MATRIX ---
cm = confusion_matrix(test_labels, pred_classes)
df_cm = pd.DataFrame(cm, index=classes, columns=classes)

plt.figure(figsize=(10, 8))
sns.heatmap(df_cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
